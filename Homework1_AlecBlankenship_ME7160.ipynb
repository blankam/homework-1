$$r_1 = -\frac{1}{2}+\frac{\sqrt{401}}{2}$$

## 2.G

## 3.A
The equations for the system in 3.A is given as follows:

## $$\ddot{x}+x=0$$

for this representation, the "velocity" or $\dot{x}$ is just equal to the negative "position" or $x$ as shown in the equation here:

A quiver plot is given below to represent the phase (couldnt figure out how to get $\dot{x} vs x$ for the first order system. The quiver plot shows an equilibrium at x=0

The ODE for this system yields the following state space representation

## $$\dot{x_1} = x_2$$
## $$\dot{x_2} = \frac{-1}{10}x_2-10x_1$$

## $$\ddot{x}+x-0.5x^{3}=0$$

Quiver plot was given below as the phase portrait. As the value for $x$ approaches 0 the trajectory will go towars the right of the plot asymptotically. 


```python
pylab.rcParams['figure.figsize'] = 20, 12
"""Problem 3A"""
def sdof_soln(x = np.asarray([1,2])):
    x1_point = x[1,:]/x[1,:]
    x2_point = x[0,:]
    return x1_point, x2_point

numx = 50
numv = 50
x = np.linspace(-10, 10, numx)
v = np.linspace(-10, 50, numv)
x0, v0 = np.meshgrid(x, v)
x_all = np.asarray([x,v])

u_1, u_2 = sdof_soln(x=x_all)
plt.xlabel('$x$', fontsize=30)
plt.ylabel('$\dot{x}$', fontsize=30)
plt.grid()
plt.quiver(x0,v0,u_1,u_2,scale = 100)
plt.show()

"""Problem 3B"""

def Problem3B(max_time=10.0, x0 = np.array([[1,1,-1,-1]]).T, v0 = np.array([[1,-1,1,-1]]).T, plotnow = 1):
    def deriv(x1_x2, t):
        """Compute the time-derivative of a SDOF system."""
        x1, x2 = x1_x2
        return [x2, (-1/10)*x2-10*x1]
    x0 = np.concatenate((x0, v0), axis = 1)
    N = x0.shape[0]
    t = np.linspace(0, max_time, int(250*max_time))
    x_t = np.asarray([sp.odeint(deriv, x0i, t)
                      for x0i in x0])
    if plotnow == 1:
        colors = plt.cm.jet(np.linspace(0, 1, N))
        fig = plt.figure()
        for i in range(N):
            x, v = x_t[i,:,:].T

            plt.plot(x, v)
        plt.xlabel('$x$',fontsize = 30)
        plt.ylabel('$\dot{x}$',fontsize = 30)
        plt.grid()

    return t, x_t
x0 = np.random.random((5,1))*2-1
v0 = np.random.random((5,1))*2-1
_, _ = Problem3B(x0 = x0, v0 = v0)

def sdof_soln(u = np.asarray([1,2])):
    u1_point = u[1,:]
    u2_point = (-1/10)*u[1,:]-10*u[0,:]
    return u1_point, u2_point

numx = 35
numv = 35
u1all = np.linspace(-1, 1, numx)
u2all = np.linspace(-3, 3, numv)
u1, u2 = np.meshgrid(u1all, u2all)
u1.shape = (numx*numv,1)
u2.shape = (numx*numv,1)
u = np.asarray([u1,u2])
u1, u2 = np.meshgrid(u1all, u2all)
u_1, u_2 = sdof_soln(u=u)
u_1 = u_1.reshape(numx, numv)
u_2 = u_2.reshape(numx, numv)
plt.grid()
plt.quiver(u1,u2,u_1,u_2)
plt.show()

"""Problem 3C"""
def sdof_soln(x = np.asarray([1,2])):
    x1_point = x[1,:]/x[1,:]
    x2_point = -x[0,:]+0.5*x[0,:]**3
    return x1_point, x2_point

numx = 50
numv = 50
x = np.linspace(-10, 10, numx)
v = np.linspace(-10, 50, numv)
x0, v0 = np.meshgrid(x, v)
x_all = np.asarray([x,v])

u_1, u_2 = sdof_soln(x=x_all)
plt.xlabel('$x$', fontsize=30)
plt.ylabel('$\dot{x}$', fontsize=30)
plt.grid()
plt.quiver(x0,v0,u_1,u_2,scale = 100)
plt.show()
```

### 4 - Plot Phase Portrait with varying $\mu$

The system is described by the equation:

## $$\ddot{x}=-\mu(x^2-1)\dot{x}-x$$

The state space representaton of this system is given by the following system of first order equations:

## $$\dot{x_1} = x_2$$
## $$\dot{x_2} = -\mu(x_1^{2}-1)x_2-x_1$$

The phase portrait is given in the next cell.

The portrait shows:

if $\mu$ is negative the trajectories will collapse into a singular point. The greater the negative value is, the quicker the trajectories collapse

if $\mu$ is equal to zero, the trajectories will fall into individual orbits  without converging on a single solution

if $\mu$ is positive, the trajectories will move out int a single orbit in which all trajectories will end up. The greater the value for $\mu$, the larger the orbit becomes

```python
list = [-10,-1,0,1,10]
for mu in list:  
    def Problem4(max_time=10.0, mu = 1, x0 = np.array([[1,1,-1,-1]]).T, v0 = np.array([[1,-1,1,-1]]).T, plotnow = 1):
        def deriv(x1_x2, t):
            """Compute the time-derivative of a SDOF system."""
            x1, x2 = x1_x2
            return [x2, -mu*(x1**2-1)*x2-x1]
        x0 = np.concatenate((x0, v0), axis = 1)
        N = x0.shape[0]
        t = np.linspace(0, max_time, int(250*max_time))
        x_t = np.asarray([sp.odeint(deriv, x0i, t)
                          for x0i in x0])
        if plotnow == 1:
            colors = plt.cm.jet(np.linspace(0, 1, N))
            fig = plt.figure()
            for i in range(N):
                x, v = x_t[i,:,:].T

                plt.plot(x, v)
            plt.xlabel('$x$',fontsize = 30)
            plt.ylabel('$\dot{x}$',fontsize = 30)
            A = str(mu)
            strin = "Problem 4, case mu ="
            B = strin+A
            plt.title(B,fontsize = 30)
            plt.grid()

        return t, x_t
    x0 = np.random.random((10,1))*2-1
    v0 = np.random.random((10,1))*2-1
    _, _ = Problem4(max_time=30,mu = mu, x0 = x0, v0 = v0)

    def sdof_soln(u = np.asarray([1,2])):
        u1_point = u[1,:]
        u2_point = (-1/10)*u[1,:]-10*u[0,:]
        return u1_point, u2_point

    numx = 35
    numv = 35
    u1all = np.linspace(-3, 3, numx)
    u2all = np.linspace(-3, 3, numv)
    u1, u2 = np.meshgrid(u1all, u2all)
    u1.shape = (numx*numv,1)
    u2.shape = (numx*numv,1)
    u = np.asarray([u1,u2])
    u1, u2 = np.meshgrid(u1all, u2all)
    u_1, u_2 = sdof_soln(u=u)
    u_1 = u_1.reshape(numx, numv)
    u_2 = u_2.reshape(numx, numv)
    plt.grid()
    plt.quiver(u1,u2,u_1,u_2)
    plt.show()
```
